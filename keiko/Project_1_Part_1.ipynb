```python
all_states_history_path = "data/all-states-history.csv"
national_history_path = "data/national-history.csv"

all_states_history = pd.read_csv(all_states_history_path)
national_history = pd.read_csv(national_history_path)

#split date column
all_states_history[["year", "month", "day"]] = all_states_history["date"].str.split('-',expand=True)

#make new df
all_states_df = pd.DataFrame(all_states_history[["date", "year", "month", "state", "death", "positive", "recovered", "totalTestResults"]])


#remove NaN from columns
all_states_df["death"] = all_states_df["death"].fillna(0)
all_states_df["positive"] = all_states_df["positive"].fillna(0)
all_states_df["recovered"] = all_states_df["recovered"].fillna(0)
all_states_df["totalTestResults"] = all_states_df["totalTestResults"].fillna(0)

#adding percent infected column
#add percent sign 
all_states_df["Percent Infected Rate"] = (all_states_df["positive"]/all_states_df["totalTestResults"])

all_states_df.head()
```
