```python
from pytrends.request import TrendReq

pytrends = TrendReq(hl='en-US', tz=360)
```

```python
kw_list = ['_']
cat_list = [3, 184, 316]
for cat in cat_list:
    pytrends.build_payload(kw_list, cat=cat, timeframe='2019-01-01 2020-12-20', geo='US-GA')
    interest_by_region_df = pytrends.interest_by_region(kw_list)
    print(interest_by_region_df)
```
