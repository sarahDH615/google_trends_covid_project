```python
# import dependencies, and pytrends
from pytrends.request import TrendReq
import pandas as pd
import time
import numpy as np

pytrends = TrendReq()
```

```python
#reading in the csv of the state abbreviations
states_path = 'statesAbbrev.csv'
states_abbrev_df = pd.read_csv(states_path)
states_abbrev_df.head()
```

```python
#creating a list of the state abbreviations with the country code added in
abbrev_list = [f"US-{x}" for x in states_abbrev_df['Code']]
len(abbrev_list)
```

```python
state_names_list = [x for x in states_abbrev_df['State']]
```

```python
states_and_abbrevs_dict = {
    'state_name': state_names_list,
    'state_abbrev': abbrev_list
}
```

```python
states_and_abbrevs_df = pd.DataFrame(states_and_abbrevs_dict)
```

```python
output_path = 'statesUSAbbrevs.csv'
states_and_abbrevs_df.to_csv(output_path, index=False)
```

```python
#attempting process per category instead of per state
#creating list of columns to go into each category's df
state_columns_list = ['time'] + abbrev_list
```

```python
path = 'top_cats.csv'
top_cats_df = pd.read_csv(path)
top_cats_df
```

```python
cat_names_list = [x for x in top_cats_df['cat_name']]
cat_index_list = [x for x in top_cats_df['cat_index']]
```

```python
#creating a dict to hold all the different top category info for all the states
top_cat_info_df_dict = {}
for cat in cat_names_list:
    top_cat_info_df_dict[cat] = pd.DataFrame(columns=state_columns_list)
```

```python
#test printing a df
top_cat_info_df_dict['Sports News']
```

```python
len(top_cat_info_df_dict)
```

```python
kw_list = '*'

def per_top_cat_info_collector(category_index_num):
    for x in range(0, len(abbrev_list)):
        try:
            pytrends.build_payload(kw_list, cat=cat_index_list[category_index_num], timeframe='2019-01-01 2020-12-20', geo=abbrev_list[x])
            cat_interest_df = pd.DataFrame(pytrends.interest_over_time())
            cat_interest_df.reset_index(level=0, inplace=True)
            top_cat_info_df_dict[cat_names_list[category_index_num]]['time'] = cat_interest_df['date']
            top_cat_info_df_dict[cat_names_list[category_index_num]][f'{abbrev_list[x]}'] = cat_interest_df['*']
            time.sleep(6)
        except (IndexError, KeyError):
            print(f"problem with {cat_names_list[category_index_num]}")
            continue
    print(top_cat_info_df_dict[cat_names_list[category_index_num]])
    top_cat_info_df_dict[cat_names_list[category_index_num]]
    output_path = f'google_trends_csvs/top_trends/{cat_names_list[category_index_num]}_states.csv'
    top_cat_info_df_dict[cat_names_list[category_index_num]].to_csv(output_path, index=False)
```

```python
for x in range(0, 4):
    per_top_cat_info_collector(x)
```

```python
path = 'alt_top_cats.csv'
alt_top_cats_df = pd.read_csv(path)
alt_top_cats_df
```

```python
alt_cat_names_list = [x for x in alt_top_cats_df['alt_cat_name']]
alt_cat_index_list = [x for x in alt_top_cats_df['alt_cat_index']]
```

```python
alt_top_cat_info_df_dict = {}
for cat in alt_cat_names_list:
    alt_top_cat_info_df_dict[cat] = pd.DataFrame(columns=state_columns_list)
```

```python
kw_list = '*'

def alt_per_top_cat_info_collector(category_index_num):
    for x in range(0, len(abbrev_list)):
        try:
            pytrends.build_payload(kw_list, cat=alt_cat_index_list[category_index_num], timeframe='2019-01-01 2020-12-20', geo=abbrev_list[x])
            cat_interest_df = pd.DataFrame(pytrends.interest_over_time())
            cat_interest_df.reset_index(level=0, inplace=True)
            alt_top_cat_info_df_dict[alt_cat_names_list[category_index_num]]['time'] = cat_interest_df['date']
            alt_top_cat_info_df_dict[alt_cat_names_list[category_index_num]][f'{abbrev_list[x]}'] = cat_interest_df['*']
            time.sleep(10)
        except (IndexError, KeyError):
            print(f"problem with {alt_cat_names_list[category_index_num]}")
            continue
    print(alt_top_cat_info_df_dict[alt_cat_names_list[category_index_num]])
    alt_top_cat_info_df_dict[alt_cat_names_list[category_index_num]]
    output_path = f'google_trends_csvs/top_trends/{alt_cat_names_list[category_index_num]}_states.csv'
    alt_top_cat_info_df_dict[alt_cat_names_list[category_index_num]].to_csv(output_path, index=False)
```

```python
for x in range(0, 4):
    alt_per_top_cat_info_collector(x)
```
